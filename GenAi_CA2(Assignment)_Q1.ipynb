```python
import pandas as pd
import numpy as np
```

```python
# Set Pandas to display all rows
pd.set_option('display.max_rows', None)

# Set seed for reproducibility
np.random.seed(42)

# Parameters
num_accounts = 100
initial_balance_range = (1000, 10000)

# Generate random account balances and transactions
data = {
    'AccountID': np.arange(1, num_accounts + 1),
    'InitialBalance': [round(x, 2) for x in np.random.uniform(1000, 10000, 100)],
    'FinalBalance': [round(x, 2) for x in np.random.uniform(500, 9000, 100)]
}

# Create DataFrame
df_sorted = pd.DataFrame(data)

# Sort accounts by FinalBalance (lowest to highest)
df_sorted = df_sorted.sort_values(by='FinalBalance')

# Display all rows
print(df_sorted[['AccountID', 'InitialBalance', 'FinalBalance']])
```
